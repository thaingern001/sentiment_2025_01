```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
```

```python
sentences = "Hello there. How are you doing today? I hope you're having a great day!"
tokens = nltk.word_tokenize(sentences)
print(tokens)
```

```python
tagged = nltk.pos_tag(tokens,lang='eng')
print(tagged)
```

```python
my_tokens = ['I', 'love', 'programming', 'in', 'Python', '.','black','cats','are','awesome','Jason','Kevin']
my_tagged = nltk.pos_tag(my_tokens, lang='eng')
print(my_tagged)
```
